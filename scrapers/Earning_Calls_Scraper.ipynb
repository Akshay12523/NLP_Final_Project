```python
# testing list comprehension
ticker = [t.split('/')[6] for t in l]
```

```python
# finding href links
tail = []
ti = []

for x in l:
    
    response = requests.get(x)
    page = response.text
    soup = BeautifulSoup(page, 'html.parser')
    href = re.findall(r'/earnings/[\w\s|\-|0-9|/]+', str(soup)) # looks for all earning call links
    for h in href: # because it returns a list of links
        tail.append(h) # iterate through each list, append each link to tail list
        
    for t in range(len(href)):
        ti.append(x.split('/')[6]) # Append ticker name by len of href

tail
```

```python
for x, y in zip(ti, tail):
    print(x, y)
```

```python
master = []
idx = []

# Check for duplicates
for n, link in enumerate(tail):
    if link not in master:
        master.append(link)
        idx.append(n)
    else:
        pass
    
ti2 = [ti[i] for i in idx]
```

```python
for x,y in zip(ti2, master):
    print(x,y)
```

# Iterate over href links

```python
# empty list to append to
transcript = []

# for each href link
for link in master:
    x = head + link + '.aspx'
    response = requests.get(x)
    print(x)
    #print(response)
    page = response.text
   
    soup = BeautifulSoup(page, 'html.parser')
    
    all_matches = soup.find("div",{"class":"main-col"}) # Looks for the transcript section
    call = (all_matches.get_text()).strip('\n') # removes new lines
    call = call.splitlines() # convert to line to prepare for lazy edits
    call = call[29:-23] # lazy edits
    transcript.append(call)
    #print(transcript)
```

```python
# create dataframe
dic = {'href': master,
      'Transcript': transcript,
      'Ticker': ti2}

df=pd.DataFrame.from_dict(dic)
df
```
