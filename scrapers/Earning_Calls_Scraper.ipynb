```python
import os #directories
import pandas as pd # dataframe
import re # regular expressions

from bs4 import BeautifulSoup # webscrape
import urllib # webscrape
import requests # webscrape
from urllib.request import urlopen # webscrape
import csv # csv?
```

# Import all transcript links

```python
head = 'https://www.fool.com'

# List of urls to iterate over. 
with open(r'C:\Users\Ryan\Documents\GitHub\NLP_Final_Project\data\transcript_list.txt', 'r') as f:
    l = [line.strip() for line in f]
    
l[:5]
```

```python
tail = [] # tail of the transcript link
ti = [] # list of tickers

# Loop of the links of earning calls transcript page
for x in l:
    
    response = requests.get(x)
    page = response.text
    soup = BeautifulSoup(page, 'html.parser')
    href = re.findall(r'/earnings/[\w\s|\-|0-9|/]+', str(soup)) # looks for all earning call links
    
    for h in href: # because it returns a list of links
        tail.append(h) # iterate through each list, append each link to tail list
        
    for t in range(len(href)): # for the length of the href list
        ti.append(x.split('/')[6]) # Append ticker name
```

```python
# print ticker , tail 
for x, y in zip(ti, tail):
    print(x, y)
```

```python
# Check for duplicates

master = [] # list of unique tail links list
idx = [] # index in which duplicate does not appear list

# Iterate over the tail list
for n, link in enumerate(tail):
    if link not in master:
        master.append(link)
        idx.append(n)
    else:
        pass

# list of tickers corresponding with tail link
ti2 = [ti[i] for i in idx]
```

```python
# Print ticker name along with tail link. 
for x,y in zip(ti2, master):
    print(x,y)
```

# Iterate over href links

```python
# empty list to append to
transcript = []

# for each href link
for link in master:
    x = head + link + '.aspx' # combine header + transcript tail link
    response = requests.get(x)
    print(x)
    #print(response)
    page = response.text
   
    soup = BeautifulSoup(page, 'html.parser')
    
    all_matches = soup.find("div",{"class":"main-col"}) # Looks for the transcript section
    call = (all_matches.get_text()).strip('\n') # removes new lines
    call = call.splitlines() # convert to line to prepare for lazy edits
    call = call[29:-23] # lazy edits
    transcript.append(call)
```

```python
# create dataframe
dic = {'href': master,
      'Transcript': transcript,
      'Ticker': ti2}

df=pd.DataFrame.from_dict(dic)
df
```

```python
# Write to csv
df.to_csv(r'C:\Users\Ryan\Documents\GitHub\NLP_Final_Project\data\earning_calls.csv', index = None, header = True)
```
